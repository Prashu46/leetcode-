```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        minprice=999
        max_profit=0
        for price in prices:
            minprice=min(minprice,price)
            max_profit=max(max_profit,price-minprice)

        return max_profit
```

```python
class Solution:
    def isIsomorphic(self, s: str, t: str) -> bool:
        hash1,hash2={},{}
        if len(s)!=len(t):
            return False
        for i in  range (len(s)):
            a1,b1=s[i],t[i]
            if ((a1 in hash1 and hash1[a1]!=b1) or (b1 in hash2 and hash2[b1]!=a1)):
                return False
            hash1[a1]=b1
            hash2[b1]=a1
        return True
```
