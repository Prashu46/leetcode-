```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        minprice=999
        max_profit=0
        for price in prices:
            minprice=min(minprice,price)
            max_profit=max(max_profit,price-minprice)

        return max_profit
```

```python
class Solution:
    def isIsomorphic(self, s: str, t: str) -> bool:
        hash1,hash2={},{}
        if len(s)!=len(t):
            return False
        for i in  range (len(s)):
            a1,b1=s[i],t[i]
            if ((a1 in hash1 and hash1[a1]!=b1) or (b1 in hash2 and hash2[b1]!=a1)):
                return False
            hash1[a1]=b1
            hash2[b1]=a1
        return True
```

```python
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        intervals.sort(key=lambda i:i[0])
        res=[]
        n=len(intervals)
        start1=intervals[0][0]
        end1=intervals[0][1]
        for i in range (1,n):
            start2=intervals[i][0]
            end2=intervals[i][1]
            if end1>=start2:
                start1=start1
                end1=max(end1,end2)
                continue
            res.append([start1,end1])
            start1=start2
            end1=end2
        res.append([start1,end1])
        return res


```
