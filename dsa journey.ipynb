```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        minprice=999
        max_profit=0
        for price in prices:
            minprice=min(minprice,price)
            max_profit=max(max_profit,price-minprice)

        return max_profit
```

```python
s=Solution()
s.maxProfit([7,1,5,3,6,4])
```
